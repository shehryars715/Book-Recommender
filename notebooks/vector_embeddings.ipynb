```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
import pickle
import pandas as pd
```

```python
# Load your CSV file
df = pd.read_csv("merged_category_dataset.csv")

df["isbn13_description"] = df["isbn13"].astype(str) + " " + df["description"].fillna("")
texts = df["isbn13_description"].tolist()
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
embeddings = model.encode(texts, convert_to_numpy=True, show_progress_bar=True)
```

```python
index = faiss.IndexFlatL2(384)  # L2 distance
index.add(embeddings)  # add vectors to index
```

```python
faiss.write_index(index, "vectors.faiss")

with open("metadata.pkl", "wb") as f:
    pickle.dump(texts, f)
```
